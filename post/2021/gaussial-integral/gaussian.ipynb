```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
from scipy.interpolate import griddata
```

```python
def gaussian(arr_x, arr_y):
    px, py, pz  = [], [], []
    for x in arr_x:
        for y in arr_y:
            px.append(x)
            py.append(y)
            pz.append(np.exp(-(x**2 + y**2)))
    df = pd.DataFrame({'x': px, 'y': py, 'z': pz})
    return df
```

```python
pts = np.linspace(-3, 3, 100)
df = gaussian(pts, pts)
```

```python
X, Y = pts, pts
Z = df.z.to_numpy().reshape(100,100)
```

```python
fig = go.Figure()
fig.add_surface(x=X, y=Y, z=Z, colorscale='Reds_r', colorbar_thickness=25, colorbar_len=0.75)
fig.update_layout(autosize=False,
                  width=700, height=500,
                  margin=dict(l=35, r=30, b=35, t=30))
fig.show()
```

```python
fig.write_json("gaussian.json")
```
